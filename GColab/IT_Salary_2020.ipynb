ChatGPT Prompt : Cara mengubah tipe data fitur "Total years of experience",
                  "Years of experience in Germany",
                  "Yearly bonus + stocks in EUR",
                  "Annual bonus+stocks one year ago. Only answer if staying in same country",
                  "Number of vacation days" dari kategorikal menjadi numerik

```python
feature_select = ["Total years of experience",
                  "Years of experience in Germany",
                  "Yearly bonus + stocks in EUR",
                  "Annual bonus+stocks one year ago. Only answer if staying in same country",
                  "Number of vacation days"]

# Mengubah tipe data fitur pada 'feature_select' menjadi numerik
for feature in feature_select:
  data[feature] = pd.to_numeric(data[feature], errors = 'coerce')
```

ChatGPT Prompt : cara menampilkan baris yang memiliki nilai null

```python
# Membuat variabel berisi baris-baris yang memiliki nilai null
value_null = data[data.isnull().any(axis = 1)]

# Pemanggilan variabel
value_null
```

ChatGPT Prompt : Cara mengisi baris dari fitur "Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week" yang bernilai null dengan "0"

```python
data["Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week"].fillna(0, inplace=True)

data.head()
```

ChatGPT Prompt : beri kode untuk menampilkan pie diagram chart untuk fitur Age dengan min 20, max 69

```python
import matplotlib.pyplot as plt

# Data contoh untuk Age
age_data = [20, 30, 40, 50, 60, 69]
labels = ['20-30', '31-40', '41-50', '51-60', '61+']

data['Age_group'] = pd.cut(data['Age'], bins = age_data, labels = labels, right = False)

# Menghitung jumlah individu dalam masing-masing kategori umur
age_count = data['Age_group'].value_counts()

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_count, labels=age_count.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Memastikan pie chart terlihat bulat
plt.title('Distribusi Umur')
plt.show()
```
